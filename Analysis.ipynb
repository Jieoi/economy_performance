```python
import pandas as pd
import statsmodels.api as sm
```

```python
# Load the CSV file
file_path = r"clean_Data.csv"
singapore_data = pd.read_csv(file_path)

# Output of processed data
singapore_data.head()
```

```python
singapore_data.columns
```

```python
# Drop 'Unnamed: 0' column since it's not relevant
singapore_data = singapore_data.drop(columns=['Unnamed: 0'])

# Get the 'Year' column as the independent variable
X = singapore_data['Year']
X = sm.add_constant(X)  # Add a constant term to the independent variable

# Iterate over each column (excluding 'Year')
for col in singapore_data.columns:
    if col != 'Year':
        # Define the dependent variable as the current column
        y = singapore_data[col]
        
        # Perform linear regression
        model = sm.OLS(y, X).fit()
        
        # Print the summary of the regression results for each column
        print(f"Regression results for {col}:")
        print(model.summary())
        print("\n" + "="*80 + "\n")  # Divider between each regression output
```
